```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
x = np.array([1, 2, 1, 2])
plt.figure()
plt.stem(x)
plt.grid()
plt.xlabel("Samples (n)", size=14)
plt.ylabel("x[n]", size=14)
```

```python
for lag in np.arange(-3, 4):
    plt.figure()
    plt.stem(np.arange(x.size)+lag, x, markerfmt='ro', linefmt='r-', basefmt='k-', label="x[n-({})]".format(lag))
    plt.stem(np.arange(x.size), x, basefmt='k-', label="x[n]")
    plt.xlabel("Samples (n)", size=14)
    plt.ylabel("Amplitude", size=14)
    plt.legend()
    plt.savefig("/tmp/lag{}.png".format(lag))
```

```python
for lag in np.arange(-2, -4, -1): # from 3 to -4
    pad_left = lag if lag>0 else 0
    pad_right = np.abs(lag) if lag<0 else 0
    print("pad_l: ", pad_left,",  pad_r: ",pad_right)
    sig = np.pad(x, (pad_left, pad_right)) * np.pad(x, (pad_right, pad_left))
    print(sig.size)
    print(sig)
    plt.figure()
    plt.stem(np.arange(-pad_left, sig.size-pad_left), sig, basefmt='k-')
    plt.xlabel("Samples (n)", size=14)
    plt.ylabel("Amplitude", size=14)
```

```python
print(np.pad(x, (0, 3)), "\n", np.pad(x,(3, 0)))
```

```python
output = np.correlate(x,x, mode="full")
plt.figure()
plt.stem(np.arange(-3,4), output)
plt.grid()
plt.xlabel("Samples (n)", size=14)
plt.ylabel("Autocorrelation", size=14)
```
