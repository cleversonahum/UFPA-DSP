```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
n = np.arange(0,10)
func = np.power((1/2), n)
```

```python
plt.figure()
plt.stem(n, func)
plt.show()
```

```python
w = np.linspace(0, np.pi*4, 100)
y = (2*np.exp(1j*w))/(-1+2*np.exp(1j*w)) # Fourier transform from "func"
#y1 = np.fft.fft(func, n=10)*1/10
A = np.abs(y)
teta = np.angle(y, deg=True)
#print(np.abs(y1) - np.abs(y[0:]))
```

```python
plt.figure()
plt.plot(w,A)
plt.show()
```

```python
plt.figure()
plt.plot(w,teta)
plt.show()
```
