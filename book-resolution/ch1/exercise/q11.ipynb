```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
Fs =60
Ts =1/ Fs # sampling frequency Fs in Hz and period Ts in sec
f = Fs/2 # key thing : Fs should be greater than 2 f0
t =np.arange(0,1+Ts,Ts) # discrete - time axis , from 0 to 1 second
theta = np.pi # define an arbitrary angle
X = np.sin(2*np.pi*f*t + theta) # not obeying sampling theorem
```

```python
plt.figure()
plt.plot(t, X)
plt.show()
```

Using the frequency of the sinoid exactly as $Fs/2$ creates an aliasing in the signal in the higher frequency (since 0 is also sampled). So, using a sinusoid with this frequency and with a $\theta=\pi$ makes the sin value becomes 0 in all sampled points (values in the graph are very near of zero due to python representation). it proves thats the Nyquist theorem needs to be $F_s>\frac{F_{max}}{2}$
