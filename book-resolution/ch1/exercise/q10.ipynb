```python
import matplotlib.pyplot as plt 
import numpy as np
import sounddevice as sd
%matplotlib inline
```

```python
Fs = 44100
A7 = 3520
A6 = 1760
A5 = 880
A4 = 440
```

```python
# a)
omega_A7 = A7*2*np.pi/Fs
omega_A4 = A4*2*np.pi/Fs
print("Omega A7: ", omega_A7,"\nOmega A4: ",omega_A4)
```

```python
# b)
Fs = 8192 # 44100 Hz is the past value, the new to generate a 3 second audio is 8192 Hz 
W1 = A5*np.pi/Fs
W2 = A6*np.pi/Fs
Ts =1/ Fs 
N = 3/Ts
n = np.arange(0,N) 
x = np.cos(W1*n) + np.cos(W2*n)
```

```python
plt.figure()
plt.plot(n/Fs, x)
plt.show()
sd.play(x, 8192)
# The audio has 3 seconds, but the player uses a different sampling frequency
```

```python
# c)
Fs = 2000 # 2 kHz
omega_A7 = A7*2*np.pi/Fs
omega_A4 = A4*2*np.pi/Fs
W1 = A5*np.pi/Fs
W2 = A6*np.pi/Fs
Ts =1/ Fs 
N = 3/Ts
n = np.arange(0,N) 
x = np.cos(W1*n) + np.cos(W2*n)
print("Omega A7: ", omega_A7,"\nOmega A4: ",omega_A4)
plt.figure()
plt.plot(n/Fs, x)
plt.show()
sd.play(x, 8192)
```
