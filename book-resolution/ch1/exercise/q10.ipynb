```python
import matplotlib.pyplot as plt 
import numpy as np
import sounddevice as sd
%matplotlib inline
```

```python
Fs = 8192 # 44100 Hz is the past value, the new to generate a 3 second audio is 8192 Hz 
W1 = 3520*np.pi/Fs
W2 = 1760*np.pi/Fs
Ts =1/ Fs 
N = 3/Ts
n = np.arange(0,N) 
x = np.cos(W1*n) + np.cos(W2*n)
```

```python
plt.figure()
plt.plot(n/Fs, x)
plt.show()
# The audio has 3 seconds, but the player uses a different sampling frequency
```

```python
sd.play(x, 8192)
```
