```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
mean = 4
variance = 3
samples = 100
random1 = np.sqrt(variance) * np.random.randn(samples) + mean
print("Mean: ",np.mean(random1), "\nVariance: ",np.var(random1))
```

```python
plt.figure()
plt.hist(random1, 20)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
# Using the formula which defines the variance of a uniform distibution as `variance = (b-a)² / 12` we have that to obtain a variance of 7 the distance between a and b needs to be of sqrt(84)
rand_temp = np.random.uniform(-np.sqrt(84)/2,np.sqrt(84)/2,samples) + 5 # 5 is the mean
plt.figure()
plt.hist(rand_temp, 20)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Mean: ",np.mean(rand_temp), "\nVariance: ",np.var(rand_temp))
```

```python
# Creating the random signal with independent but not identically distributed
random2 = np.zeros(samples*2)
random2[0::2] = random1
random2[1::2] = rand_temp
```

```python
plt.figure()
plt.hist(random2, 20)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
print("Mean: ",np.mean(random2), "\nVariance: ",np.var(random2))
```
