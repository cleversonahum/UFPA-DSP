```python
import matplotlib.pyplot as plt 
import numpy as np
%matplotlib inline
```

```python
mean = 4
variance = 3
samples = 100
random1 = np.sqrt(variance) * np.random.randn(samples) + mean
```

```python
plt.figure()
plt.hist(random1, 20)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
rand_temp = 7*np.random.uniform(-1,1,1000000)+5
plt.figure()
plt.hist(rand_temp, 20)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```
